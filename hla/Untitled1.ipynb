```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
n = 10
k = 5
m = 3
```

```python
def prob(n, m, k, 位):
    p = 1 - sp.stats.binom(m, 位).pmf(0)
    return sp.stats.binom(n, p).pmf(k)
```

```python
ps = [prob(n, m, k, 位) for 位 in np.arange(0,1, 0.0001)]
```

```python
plt.plot(np.arange(0,1, 0.0001), ps)
```
