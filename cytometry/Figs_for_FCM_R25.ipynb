```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 100)
y = np.arcsinh(x)
```

```python
np.arcsinh(1)
```

```python
plt.plot(x, y)
plt.title('Inverse hyperbolic sinh transform')
```

```python
import joypy
```

```python
x1 = np.concatenate([np.random.normal(0,1,100), np.random.normal(13, 1,10)])
x2 = np.concatenate([np.random.normal(2,1,80), np.random.normal(18, 1,15)])
x3 = np.concatenate([np.random.normal(5,1,120), np.random.normal(28, 1,20)])
```

```python
joypy.joyplot([x1, x2, x3],
              figsize=(4,4),
              colormap=plt.cm.autumn_r, 
              overlap=1.5)
pass
```
