```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x0, λ, t):
    return x0*np.exp(-λ*t)
```

```python
t = np.linspace(0, 4, 100)
```

```python
x1 = 10
x2 = 20
λ1 = 1
λ2 = 2

y1 = f(x1, λ1, t)
y2 = f(x2, λ2, t)
```

```python
y = y1 + y2
```

```python
plt.plot(t, y1, 'b-')
plt.plot(t, y2, 'g-')
plt.plot(t, y, 'r-')
plt.legend([f'λ = {λ1} x0={x1}', f'λ = {λ2} x0={x2}', f'Mixture'])
pass
plt.savefig('mix_exp.png')
```
